1. Which team won the most matches?

````python
match_wins = df['match_winner'].value_counts()
match_wins
sns.barplot(x=match_wins.values,y=match_wins.index)
plt.title("Most match win by teams")
````

2. Toss Decision Trends

````python
sns.countplot(x=df['toss_decision'])
plt.title("Toss Trend")
````

3. Toss Winner vs Match Winner

````python
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percent = count/df.shape[0]*100
percent
````

4. How do teams win? (Runs vs Wickets)

````python
sns.countplot(x=df['won_by'])
````

### Key Player Performances

1. Most Player of the match awards

````python
count = df['player_of_the_match'].value_counts().head(10)
sns.barplot(x=count.values,y=count.index)
plt.xlabel("POM Awards")
plt.ylabel("POM Awardee")
plt.title("Player of the Match Awards")
````

2. Top Scorers

````python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=high.values,y=high.index)
plt.xlabel("Runs")
plt.ylabel("Top Scorer")
````

3. Best Bowling Figures

````python
df['highest_wickets']=df['best_bowling_figure'].apply(lambda x:x.split('--')[0])
df['highest_wickets']=df['highest_wickets'].astype(int)
````

````python
top_bowlers= df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_bowlers.values,y=top_bowlers.index)
````

### Venue Analysis

Most Matches Played by Venue

````python
venues = df['venue'].value_counts()
sns.barplot(x=venues.values,y=venues.index)
plt.xlabel("Matches Played")
plt.ylabel("Venues")
````

Who won the highest margin by runs?

````python
won_runs = df[df['won_by']=='Runs'].sort_values(by='margin',ascending=False).head(1)[['match_winner','margin']]
won_runs
````

Which player has the highest individual score?

````python
df[df['highscore']==df['highscore'].max()][['top_scorer','highscore']]
````
