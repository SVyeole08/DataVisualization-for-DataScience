```python
import seaborn as sns
import matplotlib.pyplot as plt
```

We use distribution plots to visualize the distribution of quantitative data

```python
# Seaborn itself contains multiple data sets:
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')  #Accessing the dataset
df
```

**Distribution Plot**

```python
sns.displot(df['total_bill'],bins=20,kde=True)  #displot and histplot are same
```

```python
sns.histplot(df['total_bill'],bins=20,kde=True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

**JointPlot**

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind="scatter")
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind="hist")
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind="hex")
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind="kde")
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind="reg")
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind="resid")
```

**Pairplot**

```python
#Pairplot compares all the quantitative quantities with each other and visualize them in plot
sns.pairplot(df)
```

```python
#Pairplot compares all the quantitative quantities with each other and visualize them in plot
sns.pairplot(df,hue='sex')  #Hue represents qualitative quantities in comparison
```

```python
sns.pairplot(df,hue='day')
```

```python
sns.pairplot(df,hue='time')
```

```python
sns.pairplot(df,hue='smoker')
```

**Rugplot**

```python
sns.rugplot(df)  #Rugplot not much important and not used much
```
