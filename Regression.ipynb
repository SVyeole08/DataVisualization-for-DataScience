##### lmplot()
regression plot is used in machine learning so we will not fully learn what, how and where it is used but we will see how it is created and what are the values we can put inside it

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)  #This is how we call lmplot 
```
